```python
import pandas as pd
matches = pd.read_csv('../data/raw/results.csv')
print(matches)
```

```python
print(type(matches))
print(matches.shape)
```

```python
equipe_domicile=matches['home_team']
print(type(equipe_domicile))
print(equipe_domicile.head(10))
print(equipe_domicile.nunique())
```

```python
matches_france_domicile = matches[matches['home_team']=='France']
print(matches_france_domicile.shape)
print(matches_france_domicile.head(5))
```

```python
matches_total_france = (matches[ (matches['home_team']=='France') | (matches['away_team']=='France') ] )
print(matches_total_france.shape)
```

```python
print(matches_total_france.sort_values('date',ascending=False).head(5))
print(matches.value_counts('tournament'))
```

```python
matches['date'] = pd.to_datetime(matches['date'])
matches['year']=matches['date'].dt.year
matchs_par_annee = matches.groupby('year').size()
print(matchs_par_annee.tail(10))
print(matchs_par_annee.head(10))
```

```python
matches_total_france_trie= matches_total_france.sort_values('date',ascending=True)
print(matches_total_france_trie[matches_total_france_trie['date']<'2020-01-01'].tail(1))

```
